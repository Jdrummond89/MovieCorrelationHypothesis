```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Read in the data w/ a dataframe
df = pd.read_csv(r'C:\Users\pc-47\Downloads\Movies Dataset w Excel\movies.csv')
```

```python
#Double check data
df.head()
```

```python
# Check for missing data w/ for loop 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop data that is missing or incomplete 

df = df.dropna()
```

```python
# Check data types used in this data

df.dtypes
```

```python
# Simple data cleaning; change a few data types from floats to integers 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
# Sort by gross revenue 

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Allow yourself to see all the data if you need it

pd.set_option('display.max_rows', None)
```

```python
# Build out a scatter plot with budget vs. gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs. Gross Revenue')

plt.xlabel('Gross Revenue')

plt.ylabel('Total Budget Spent')

plt.show()
```

```python
# Build a regression plot for budget vs. revenue using seaborn 

sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})

plt.title('Budget vs. Gross Revenue')

plt.xlabel('Gross Revenue')

plt.ylabel('Total Budget Spent')
```

```python
# Build in correlations 
# Pearson, Kendall, Spearman 
# Results showing high correlation between budget and gross (74%), also between budget and score (72%), but lower for budget and votes

df.corr(method='pearson')
```

```python
# Build in a correlation matrix w/ annotations  

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True) 

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Data')

plt.ylabel('Movie Data')

plt.show()
```

```python
# Convert all objects into a random numeric (cat.codes) so we can include in our correlation data
# Objects are our company/production house names, directors, country, name of movie 

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Adding in all data now to check for further correlations (like does 21st century fox correlate to high gross rev)

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True) 

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Data')

plt.ylabel('Movie Data')

plt.show()
```

```python
# Run all data so you can unstack in next function 

pd.set_option('display.max_rows', None)
```

```python
# Unstacking data to check for correlations w/out visuals 

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Or you can call a sorted values function

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Filter sorted_values by > 0.5 in a  new variable 

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
